```python
from sklearn.model_selection import train_test_split

x_train, x_val, y_train, y_val = train_test_split(
    x,
    y,
    test_size=0.2,
    random_state=42,
    stratify=y
)
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler, OneHotEncoder, OrdinalEncoder
from sklearn.pipeline import Pipeline

# 1) Define column groups
numeric_cols = [
    'age',
    'alcohol_consumption_per_week',
    'physical_activity_minutes_per_week',
    'diet_score',
    'sleep_hours_per_day',
    'screen_time_hours_per_day',
    'bmi',
    'waist_to_hip_ratio',
    'systolic_bp',
    'diastolic_bp',
    'heart_rate',
    'cholesterol_total',
    'hdl_cholesterol',
    'ldl_cholesterol',
    'triglycerides'
]

ordinal_cols = ['education_level', 'income_level']

nominal_cols = [
    'gender',
    'ethnicity',
    'smoking_status',
    'employment_status'
]

binary_cols = [
    'family_history_diabetes',
    'hypertension_history',
    'cardiovascular_history'
]

# 2) Define encoders
education_order = ['No formal', 'Highschool', 'Graduate', 'Postgraduate']
income_order = ['Low', 'Lower-Middle', 'Middle', 'Upper-Middle', 'High']

ordinal_encoder = OrdinalEncoder(
    categories=[education_order, income_order]
)

# 3) Build ColumnTransformer
preprocessor = ColumnTransformer(
    transformers=[
        ('num', StandardScaler(), numeric_cols),
        ('ord', ordinal_encoder, ordinal_cols),
        ('nom', OneHotEncoder(handle_unknown='ignore', sparse_output=False), nominal_cols),
        ('bin', 'passthrough', binary_cols)
    ],
    remainder='drop'
)
```

```python
# Fit on training data only (prevents leakage)
x_train_processed = preprocessor.fit_transform(x_train)

# Transform validation using the same fitted preprocessing
x_val_processed = preprocessor.transform(x_val)

# Quick checks
print("x_train_processed shape:", x_train_processed.shape)
print("x_val_processed shape:", x_val_processed.shape)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import roc_auc_score, accuracy_score, classification_report

# Baseline model
baseline_model = LogisticRegression(max_iter=1000, n_jobs=-1)

# Train
baseline_model.fit(x_train_processed, y_train)

# Predict probabilities + classes
y_val_proba = baseline_model.predict_proba(x_val_processed)[:, 1]
y_val_pred = (y_val_proba >= 0.5).astype(int)

# Evaluate
print("Accuracy:", accuracy_score(y_val, y_val_pred))
print("ROC-AUC:", roc_auc_score(y_val, y_val_proba))
print("\nClassification Report:\n", classification_report(y_val, y_val_pred))
```

```python
import numpy as np
from sklearn.metrics import precision_score, recall_score, roc_auc_score

thresholds = [0.3, 0.4, 0.5, 0.6, 0.7]

print("Threshold | Precision (1) | Recall (1)")
print("--------------------------------------")

for t in thresholds:
    y_pred_t = (y_val_proba >= t).astype(int)
    p = precision_score(y_val, y_pred_t)
    r = recall_score(y_val, y_pred_t)
    print(f"{t:>9} | {p:>13.3f} | {r:>9.3f}")

print("\nROC-AUC (unchanged):", roc_auc_score(y_val, y_val_proba))
```

```python
import numpy as np
import pandas as pd
import lightgbm as lgb
from lightgbm import LGBMClassifier
from sklearn.metrics import roc_auc_score

# 1) Load test and transform once (no fit)
test_df = pd.read_csv("B:/Diabatic Data/test.csv")
x_test_lgbm = lgbm_preprocessor.transform(test_df)

# 2) Prepare arrays to store predictions
oof_pred = np.zeros(x_train_lgbm.shape[0])        # out-of-fold preds for AUC check
test_pred = np.zeros(x_test_lgbm.shape[0])        # averaged test preds

# 3) LightGBM params (use your improved/generalizable settings)
base_params = dict(
    n_estimators=20000,
    learning_rate=0.02,
    num_leaves=31,
    min_child_samples=200,
    subsample=0.8,
    subsample_freq=1,
    colsample_bytree=0.8,
    reg_alpha=0.5,
    reg_lambda=1.0,
    random_state=42,
    n_jobs=-1
)

# 4) K-Fold training loop
for fold, (tr_idx, va_idx) in enumerate(kf.split(x_train_lgbm, y_train), 1):
    x_tr, x_va = x_train_lgbm[tr_idx], x_train_lgbm[va_idx]
    y_tr, y_va = y_train.iloc[tr_idx], y_train.iloc[va_idx]

    model = LGBMClassifier(**base_params)

    model.fit(
        x_tr, y_tr,
        eval_set=[(x_va, y_va)],
        eval_metric="auc",
        callbacks=[
            lgb.early_stopping(stopping_rounds=300),
            lgb.log_evaluation(period=200)
        ]
    )

    # OOF prediction for this fold
    oof_pred[va_idx] = model.predict_proba(x_va)[:, 1]

    # Test prediction (accumulate for averaging)
    test_pred += model.predict_proba(x_test_lgbm)[:, 1] / kf.get_n_splits()

    fold_auc = roc_auc_score(y_va, oof_pred[va_idx])
    print(f"Fold {fold} AUC: {fold_auc:.6f} | best_iter: {model.best_iteration_}")

# 5) Overall OOF AUC (reliable estimate)
print("\nOverall OOF ROC-AUC:", roc_auc_score(y_train, oof_pred))

# 6) Build Kaggle submission
submission_cv = pd.DataFrame({
    "id": test_df["id"],
    "diagnosed_diabetes": test_pred
})
submission_cv.to_csv("submission_cv.csv", index=False)

print("\nSaved: submission_cv.csv")
print(submission_cv.head())
```
